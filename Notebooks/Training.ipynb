```python
# Import necessary libraries
import yaml
from roboflow import Roboflow
import os
```

```python
rf = Roboflow(api_key="********") #Replace Here with you API
project = rf.workspace("segmentation-kmepg").project("tennis-7jk60")
version = project.version(2)
dataset = version.download("yolov8")


# Print the location of the downloaded dataset
dataset.location
```

# Updating YAML File Paths

```python
# Path to the YAML file
yaml_file = dataset.location + "/data.yaml"

# Read the YAML file
with open(yaml_file, 'r') as file:
    data = yaml.safe_load(file)

# Update paths
data['test'] = dataset.location + "/test/images"
data['train'] = dataset.location + "/train/images"
data['val'] = dataset.location + "/valid/images"

# Save the updated YAML file
with open(yaml_file, 'w') as file:
    yaml.safe_dump(data, file)

# Print confirmation message
print("Updated paths in the YAML file")
```
